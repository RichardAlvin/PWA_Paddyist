```python
import pandas as p 
import numpy as n  
import matplotlib. pyplot as pt  
```

```python
#Urutan item list
#UnderSampling, OverSampling
#224x224 pixel, 256x256 pixel
#EfficientNetV2B0, MobileNetV3-Large, Average Ensemble, Concatenation Ensemble
average_speed_prediction = p.Series([249,368,239,304,414,362,403,224,381,208,232,409,398,375])  
file_size_h5 = p. Series ([777,1007,577,750,977,1756,1353,777,1007,577,750,977,1756,1353])  
correlation = average_speed_prediction. corr(file_size_h5)
print (correlation)  
```

```python
#Grafik Korelasi
pt. title ('Korelasi antara Kecepatan Rata-Rata Prediksi Model \nTerhadap Ukuran Model dalam Format .h5')  
pt. scatter (average_speed_prediction, file_size_h5)  
pt. plot (n. unique (average_speed_prediction), n.poly1d (n. polyfit (average_speed_prediction, file_size_h5, 1))(n.unique (average_speed_prediction)), color = 'green')  
pt. xlabel ('Kecepatan Rata-Rata Prediksi Model (in ms)')  
pt. ylabel ('Ukuran Model dalam Format .h5 (in MB)') 
```
